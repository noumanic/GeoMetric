```python
# ── Validate processed files ───────────────────────────────
import geopandas as gpd
import pandas as pd
from scripts.utils.config import PATHS

print("\n=== Master World GeoDataFrame ===")
world = gpd.read_file(PATHS["processed"] / "master_world.gpkg")
print(f"  Countries : {len(world)}")
print(f"  Columns   : {list(world.columns)}")
print(f"  CRS       : {world.crs}")

print("\n=== Emissions ===")
emiss = pd.read_csv(PATHS["processed"] / "emissions_2020.csv")
print(f"  Records   : {len(emiss)}")
print(f"  Columns   : {list(emiss.columns)}")
print(emiss[["country","co2_per_capita","co2_total"]].dropna().head(8))

print("\n=== Airports ===")
ap = pd.read_csv(PATHS["processed"] / "airports_clean.csv")
print(f"  Airports  : {len(ap)}")
print(ap[["name","city","country","total_routes"]].nlargest(8, "total_routes"))

print("\n=== Temperature Stations ===")
tmp = pd.read_csv(PATHS["processed"] / "temperature_stations.csv")
print(f"  Stations  : {len(tmp)}")
print(tmp.head(5))
```
